```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "example1.bin"

data = np.load(file_path)

plt.imshow(data, cmap='plasma')
plt.colorbar(label="Value")
plt.title("example.bin")
plt.tight_layout()
plt.show()
```
